```python
import h5py


file_path = "D:/University_Course_work/StatsML-2/Deeplearning/Practical_Home_Work/species_learning_with_deep_learning/Data/bird_spectrograms.hdf5"

with h5py.File(file_path, 'r') as hdf:
    print("Keys in file:")
    for key in hdf.keys():
        print(f"  - {key}")

    def explore_hdf5(name, obj):
        if isinstance(obj, h5py.Dataset):
            print(f"Dataset: {name} {obj.shape}")
            print(f"  Dtype: {obj.dtype}")
            print("  Sample data:", obj[0:min(5, obj.shape[0])])
        elif isinstance(obj, h5py.Group):
            print(f"Group: {name}")

    print("\nExploring file structure and sample data:")
    hdf.visititems(explore_hdf5)
```
